```python
import IPython.display as ipd
import matplotlib.pyplot as plt
import librosa.display

wav_path = "../data/experiment_data/08_03_21_Test/"
```

```python
def visualize_wav(signal):
    plt.figure(figsize=(15,17))
    plt.subplot(3,1,1)
    librosa.display.waveplot(signal)
    plt.show()
```

```python
click = "num_2_label_29_sound_silence_20ms.wav"
sound_path = wav_path + click


signal, sr = librosa.load(sound_path)
visualize_wav(signal)

```
